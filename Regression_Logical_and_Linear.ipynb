```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Generate synthetic data for binary classification
X_binary, y_binary = make_classification(n_samples=100, n_features=3, n_informative=2, n_redundant=0, n_clusters_per_class=1, random_state=42)

# Split the data into training and testing sets
X_train_binary, X_test_binary, y_train_binary, y_test_binary = train_test_split(X_binary, y_binary, test_size=0.2, random_state=42)

# Logistic Regression for Binary Classification
logistic_reg_binary = LogisticRegression()
logistic_reg_binary.fit(X_train_binary, y_train_binary)
y_pred_binary = logistic_reg_binary.predict(X_test_binary)

# Evaluate Logistic Regression for Binary Classification
accuracy_binary = accuracy_score(y_test_binary, y_pred_binary)
confusion_matrix_binary = confusion_matrix(y_test_binary, y_pred_binary)
classification_report_binary = classification_report(y_test_binary, y_pred_binary)

print("Accuracy (Logistic Regression - Binary Classification):", accuracy_binary)
print("Confusion Matrix:\n", confusion_matrix_binary)
print("Classification Report:\n", classification_report_binary)

# Plot Binary Classification Results
plt.figure(figsize=(8, 6))
plt.scatter(X_test_binary[:, 0], X_test_binary[:, 1], c=y_test_binary, cmap=plt.cm.coolwarm, marker='o', s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Binary Classification Data')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Generate synthetic data for multiclass classification
X_multiclass, y_multiclass = make_classification(n_samples=100, n_features=3, n_informative=2, n_redundant=0, n_clusters_per_class=1, n_classes=3, random_state=42)

# Split the data into training and testing sets
X_train_multiclass, X_test_multiclass, y_train_multiclass, y_test_multiclass = train_test_split(X_multiclass, y_multiclass, test_size=0.2, random_state=42)

# Logistic Regression for Multiclass Classification
logistic_reg_multiclass = LogisticRegression(multi_class='auto')
logistic_reg_multiclass.fit(X_train_multiclass, y_train_multiclass)
y_pred_multiclass = logistic_reg_multiclass.predict(X_test_multiclass)

# Evaluate Logistic Regression for Multiclass Classification
accuracy_multiclass = accuracy_score(y_test_multiclass, y_pred_multiclass)
confusion_matrix_multiclass = confusion_matrix(y_test_multiclass, y_pred_multiclass)
classification_report_multiclass = classification_report(y_test_multiclass, y_pred_multiclass)

print("Accuracy (Logistic Regression - Multiclass Classification):", accuracy_multiclass)
print("Confusion Matrix:\n", confusion_matrix_multiclass)
print("Classification Report:\n", classification_report_multiclass)

# Plot Multiclass Classification Results
plt.figure(figsize=(8, 6))
plt.scatter(X_test_multiclass[:, 0], X_test_multiclass[:, 1], c=y_test_multiclass, cmap=plt.cm.coolwarm, marker='o', s=50)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Multiclass Classification Data')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Generate synthetic data for regression
X_regression, y_regression = make_regression(n_samples=100, n_features=1, noise=10, random_state=42)

# Split the data into training and testing sets
X_train_regression, X_test_regression, y_train_regression, y_test_regression = train_test_split(X_regression, y_regression, test_size=0.2, random_state=42)

# Linear Regression for Continuous Regression
linear_reg = LinearRegression()
linear_reg.fit(X_train_regression, y_train_regression)
y_pred_regression = linear_reg.predict(X_test_regression)

# Evaluate Linear Regression for Continuous Regression
mse_regression = mean_squared_error(y_test_regression, y_pred_regression)
r2_regression = r2_score(y_test_regression, y_pred_regression)

print("Mean Squared Error (Linear Regression - Continuous Regression):", mse_regression)
print("R-squared Score (Linear Regression - Continuous Regression):", r2_regression)

# Plot Continuous Regression Results
plt.figure(figsize=(8, 6))
plt.scatter(X_test_regression, y_test_regression, color='blue')
plt.plot(X_test_regression, y_pred_regression, color='red', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Continuous Regression Data')
plt.show()
```
